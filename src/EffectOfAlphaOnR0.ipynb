### The effect that $ R_0 $ has on $ \alpha $

Let:

- $ m $ be the mean (parameter) for the co-aggregation (negative binomial) distribution.
- $ R_0 $ be the mean (parameter) for the offspring (negative binomial) distribution.
- $ \alpha $ be the probability that a co-aggregating larva becomes an infectious nymph.

As noted above, $ R_0 = \alpha \cdot m  $

If $ R_0 \ge 1 $ then the outbreak might persist.

$ 1 \le \alpha \cdot m  $

$ \frac{1}{m} \le \alpha $

So, since $ m $ depends directly on the data, then $ \frac{1}{m} $ defines the the minimum threshold that $ \alpha $ must be for an outbreak to have a chance of persisting.

```python
import matplotlib.pyplot as plt
```

```python
# These estimates for m are from ModellingData_OffspringDistribution.ipynb
m_2004_SA_ric = 3.9286
m_2004_SA_tri = 6.3869
m_2005_FV_ric = 14.75
m_2005_SA_tri = 4.8101
m_2005_FV_tri = 0.1538

alpha_range = (0,1)
resolution = 1000

x = [float(x)/1000 for x in range(alpha_range[0],alpha_range[1]*resolution)]
x.append(1)

R0_2004_SA_ric = [i*m_2004_SA_ric for i in x]
R0_2004_SA_tri = [i*m_2004_SA_tri for i in x]
R0_2005_FV_ric = [i*m_2005_FV_ric for i in x]
R0_2005_SA_tri = [i*m_2005_SA_tri for i in x]
R0_2005_FV_tri = [i*m_2005_FV_tri for i in x]

plt.plot(x, R0_2004_SA_ric)
plt.plot(x, R0_2004_SA_tri)
plt.plot(x, R0_2005_FV_ric)
plt.plot(x, R0_2005_SA_tri)
plt.plot(x, R0_2005_FV_tri)
plt.fill_between(x,0,1, color="red", alpha=0.1)
```
